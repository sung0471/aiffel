### tf_flowers datasets
https://www.tensorflow.org/datasets/catalog/tf_flowers

```python
import tensorflow_datasets as tfds
(raw_train, raw_validation, raw_test), metadata = tfds.load(
    'tf_flowers',
    split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
    with_info=True,
    as_supervised=True,
)
```

```python
IMG_SIZE = 128 # All images will be resized to 160x160

def format_example(image, label):
    image = tf.cast(image, tf.float32)  # image=float(image)같은 타입캐스팅의  텐서플로우 버전입니다.
    image = (image/127.5) - 1
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    return image, label
```

```python
import tensorflow as tf

train = raw_train.map(format_example)
validation = raw_validation.map(format_example)
test = raw_test.map(format_example)
```

```python
BATCH_SIZE = 16
SHUFFLE_BUFFER_SIZE = 1000
train_batches = train.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
validation_batches = validation.batch(BATCH_SIZE)
test_batches = test.batch(BATCH_SIZE)
```

```python
IMG_SHAPE = (IMG_SIZE, IMG_SIZE, 3)

# # Create the base model from the pre-trained model VGG16
# base_model = tf.keras.applications.VGG16(input_shape=IMG_SHAPE,
#                                          include_top=False,
#                                          weights='imagenet')

# Create the base model from the pre-trained model VGG16
base_model = tf.keras.applications.InceptionResNetV2(input_shape=IMG_SHAPE,
                                         include_top=False,
                                         weights='imagenet')
```

```python
for image_batch, label_batch in train_batches.take(1):
    pass

image_batch.shape, label_batch.shape
```

```python
image_batch.shape
```

```python
base_model.summary()
```

```python
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
```

```python
dense_layer = tf.keras.layers.Dense(512)
batch_normalization = tf.keras.layers.BatchNormalization()
relu = tf.keras.layers.Activation('relu')
dropout = tf.keras.layers.Dropout(0.2)
prediction_layer = tf.keras.layers.Dense(5, activation='softmax')

dense_layer_2 = tf.keras.layers.Dense(512)
batch_normalization_2 = tf.keras.layers.BatchNormalization()
relu_2 = tf.keras.layers.Activation('relu')
```

```python
base_model.trainable = False
```

```python
# model = tf.keras.Sequential([
#     base_model,
#     global_average_layer,
#     dense_layer,
#     relu,
#     prediction_layer
# ])
# model = tf.keras.Sequential([
#     base_model,
#     global_average_layer,
#     dense_layer,
#     relu,
#     dropout,
#     prediction_layer
# ])
model = tf.keras.Sequential([
    base_model,
    global_average_layer,
    dense_layer,
    batch_normalization,
    relu,
    prediction_layer
])
# model = tf.keras.Sequential([
#     base_model,
#     global_average_layer,
#     dense_layer,
#     batch_normalization,
#     relu,
#     dropout,
#     prediction_layer
# ])
# model = tf.keras.Sequential([
#     base_model,
#     global_average_layer,
#     dense_layer,
#     batch_normalization,
#     relu,
#     dense_layer_2,
#     batch_normalization_2,
#     relu_2,
#     prediction_layer
# ])
```

```python
model.summary()
```

```python
# base_learning_rate = 1e-4
# model.compile(optimizer=tf.keras.optimizers.RMSprop(lr=base_learning_rate, decay=1e-4),
#               loss=tf.keras.losses.sparse_categorical_crossentropy,
#               metrics=['accuracy'])
base_learning_rate = 1e-4
model.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss=tf.keras.losses.sparse_categorical_crossentropy,
              metrics=['accuracy'])
# model.compile(optimizer=tf.keras.optimizers.Adadelta(),
#               loss=tf.keras.losses.sparse_categorical_crossentropy,
#               metrics=['accuracy'])
# base_learning_rate = 1e-4
# model.compile(optimizer=tf.keras.optimizers.Nadam(lr=base_learning_rate),
#               loss=tf.keras.losses.sparse_categorical_crossentropy,
#               metrics=['accuracy'])
```

```python
EPOCHS = 20

import os
checkpoint_dir = os.getcwd() + "/checkpoint"
callbacks = [
    tf.keras.callbacks.ModelCheckpoint(filepath=os.path.join(checkpoint_dir, '{epoch:02d}-{val_loss:.4f}.hdf5'),
                                      monitor='val_loss',
                                      verbose=1,
                                      save_best_only=True),
    tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                    patience=5)
]
history = model.fit(train_batches,
                    epochs=EPOCHS,
                    validation_data=validation_batches, callbacks=callbacks)
```

```python
import matplotlib.pyplot as plt
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

```python
for image_batch, label_batch in test_batches.take(1):
    images = image_batch
    labels = label_batch
    predictions = model.predict(image_batch)
    pass

predictions
```

```python
import numpy as np
predictions = np.argmax(predictions, axis=1)
predictions
```

```python
plt.figure(figsize=(20, 12))

for idx, (image, label, prediction) in enumerate(zip(images, labels, predictions)):
    plt.subplot(4, 4, idx+1)
    image = (image + 1) / 2
    plt.imshow(image)
    correct = label == prediction
    title = f'real: {label} / pred :{prediction}\n {correct}!'
    if not correct:
        plt.title(title, fontdict={'color': 'red'})
    else:
        plt.title(title, fontdict={'color': 'blue'})
    plt.axis('off')
```

```python
count = 0
for image, label, prediction in zip(images, labels, predictions):
    correct = label == prediction
    if correct:
        count = count + 1

print(count / BATCH_SIZE * 100) # 약 95% 내외
```

```python
def show_and_predict_image(filepath, idx, img_size=128):
    plt.subplot(1, 4, idx+1)
    image = load_img(filepath, target_size=(img_size, img_size))
    plt.imshow(image)
    plt.axis('off')
    image = img_to_array(image).reshape(1, img_size, img_size, 3)
    image = image/127.5 - 1
    prediction = model.predict(image)[0]
    
    string = 'This image seems\n | '
    label_name = ['dandelion', 'daisy', 'tulips', 'sunflowers', 'roses']
    for i in range(len(label_name)):
        string += f'{label_name[i]}: {round(prediction[i] * 100)}% | '
    
    print(string)
```

```python
   
img_list = ['my_dandelion.jpg', 'my_rose.jpeg', 'my_sunflower.jpg', 'my_tulip.jpeg']

plt.figure(figsize=(10, 15))
for i, img in enumerate(img_list):
    img_path = img_dir_path + '/' + img
    show_and_predict_image(img_path, i)
```
