```python
import pandas as pd
train_path = 'data/train.csv'
train = pd.read_csv(train_path)
print(train.shape)
train.head()
```

```python
train['datetime'] = pd.to_datetime(train['datetime'])
train['datetime']
# date = Series
```

```python
train['year'] = train['datetime'].dt.year
train['month'] = train['datetime'].dt.month
train['day'] = train['datetime'].dt.day
train['hour'] = train['datetime'].dt.hour
train['minute'] = train['datetime'].dt.minute
train['second'] = train['datetime'].dt.second
# df_date = df_date.drop('datetime', axis=1)
train
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,3)
fig.set_size_inches(18, 8)
sns.countplot(train['year'], ax=ax[0,0])
sns.countplot(train['month'], ax=ax[0,1])
sns.countplot(train['day'], ax=ax[0,2])
sns.countplot(train['hour'], ax=ax[1,0])
sns.countplot(train['minute'], ax=ax[1,1])
sns.countplot(train['second'], ax=ax[1,2])
plt.title('datetime')
plt.show()
```

```python
train.head()
```

```python
# x_col = ['season', 'holiday', 'workingday', 'weather','temp', 'atemp', 'humidity', 'windspeed', 'casual', 'registered',
#            'year', 'month', 'day', 'hour']
# x_col = ['season', 'holiday', 'workingday', 'weather','temp', 'atemp', 'humidity', 'windspeed', 'casual',
#            'year', 'month', 'day', 'hour']
x_col = ['season', 'holiday', 'workingday', 'weather','temp', 'atemp', 'humidity', 'windspeed',
           'year', 'month', 'day', 'hour']
y_col = ['count']

X = train[x_col].values
y = train[y_col].values
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=32, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train_X, train_y)
```

```python
pred_y = model.predict(test_X)
print(pred_y, test_y)
print(sum(pred_y-test_y))
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(test_y, pred_y)
rmse = mean_squared_error(test_y, pred_y, squared=False)
print(mse, rmse)
```

```python
test_X_df = pd.DataFrame(test_X, columns=x_col)
test_X_df['pred_y'] = pred_y
test_X_df['test_y'] = test_y
test_X_df
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(12, 6)
sns.lineplot(x='temp', y='pred_y', data=test_X_df, ax=ax[0])
sns.lineplot(x='temp', y='test_y', data=test_X_df, ax=ax[0])
sns.lineplot(x='humidity', y='pred_y', data=test_X_df, ax=ax[1])
sns.lineplot(x='humidity', y='test_y', data=test_X_df, ax=ax[1])
plt.show()
```

- Review

'casual'과 'registered'의 합은 count와 동일하여, 모델의 overfitting 현상에 큰 영상을 끼쳤다.   
그래서, 두 항목을 제거한 결과 overfitting을 피하면서 유사한 경향성을 보이도록 학습이 되었다.   

temp와 humidity 중에는 temp(온도)가 자전거 수 예측에 더 좋은 경향을 보였다.
