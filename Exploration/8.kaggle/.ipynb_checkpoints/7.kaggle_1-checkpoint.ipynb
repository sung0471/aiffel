##### log-scaling test

```python
xx = np.linspace(0, 10, 500)
yy = np.log(xx)

plt.hlines(0,0,10)
plt.vlines(0, -5, 5)
plt.plot(xx, yy, c='r')
plt.show()
```

- $ 0 \lt x \lt 1 $ 에서는 y에 대한 기울기가 커져, $ -\infty < y < 0 $ 범위를 가짐
    - 0 ~ 1의 값들은 y를 큰 범위로 퍼뜨려 줌   
- $ 1 \lt x $ 에서는 y에 대한 기울기가 급격히 작아짐
    - 1이상의 값들은 y를 작은 구간의 값들로 모이게 함

### 모델 설계

- blending 기법 사용 (ensemble)
    - https://subinium.github.io/introduction-to-ensemble-1/#:~:text=%EC%95%99%EC%83%81%EB%B8%94(Ensemble)%20%ED%95%99%EC%8A%B5%EC%9D%80%20%EC%97%AC%EB%9F%AC,%EB%A5%BC%20%EA%B0%80%EC%A7%80%EA%B3%A0%20%EC%9D%B4%ED%95%B4%ED%95%98%EB%A9%B4%20%EC%A2%8B%EC%8A%B5%EB%8B%88%EB%8B%A4.

- 분류기의 성능이 0.7인 5개의 분류기가 있을 때,   
  voting ensemble에서 majority 방식을 사용할 경우 예상 정확도

```python
def get_count(select):
    if select == 5 or select == 0:
        return 1
    elif select == 4 or select == 1:
        return 5
    else:
        return 10

pred = 0
select = 5
for i in range(3):
    count = get_count(select)
    pred += (0.7 ** select * 0.3 ** (5-select)) * count
    select -= 1

print(pred)
```

### Average Blending
- gboost, xgboost, lightgbm

```python
gboost = GradientBoostingRegressor(random_state=2019)
xgboost = xgb.XGBRegressor(random_state=2019)
lightgbm = lgb.LGBMRegressor(random_state=2019)

models = [{'model':gboost, 'name': 'GradientBoosting'}, 
          {'model':xgboost, 'name': 'XGBoost'},
          {'model':lightgbm, 'name': 'LightGBM'}]
```

### Cross validation

```python
def get_cv_score(models):
    kfold = KFold(n_splits=5, random_state=2019).get_n_splits(x.values)
    for m in models:
        print("Model {} CV score : {:.4f}".format(
            m['name'], np.mean(cross_val_score(m['model'], x.values,  y)), kf=kfold))
```

```python
get_cv_score(models)
```

### Make submission files

#### get prediction

```python
def AveragingBlending(models, x, y, sub_x):
    for m in models:
        m['model'].fit(x.values, y)
    
    predictions = np.column_stack([
        m['model'].predict(sub_x.values) for m in models
    ])
    
    return np.mean(predictions, axis=1)
```

```python
y_pred = AveragingBlending(models, x, y, sub)
print(len(y_pred))
y_pred
```
