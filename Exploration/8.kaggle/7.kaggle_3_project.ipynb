```python
from collections import Counter

def detect_outliers(df, n, features):
    outlier_indices = []
    for col in features: 
        Q1 = np.percentile(df[col], 25)
        Q3 = np.percentile(df[col], 75)
        IQR = Q3 - Q1
        
        outlier_step = 2.0 * IQR
        
        outlier_list_col = df[(df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)
    
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(k for k, v in outlier_indices.items() if v > n)
    
    return multiple_outliers 

# for col in train.columns:
check_col = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
       'waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long',
       'sqft_living15', 'sqft_lot15']
Outliers_to_drop = detect_outliers(train, 2, check_col)
train.loc[Outliers_to_drop]

train_drop = train.drop(Outliers_to_drop, axis = 0).reset_index(drop=True)
y_drop = y.drop(Outliers_to_drop, axis = 0).reset_index(drop=True)
```

## RMSE 함수

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(np.expm1(y_test), np.expm1(y_pred)))
```

### 모델 학습 get_score 함수

```python
def get_scores(models, train, y):
    df = {}
    for model in models:
        # 모델이름 get
        model_name = model.__class__.__name__

        # train, test 분리, random_state 고정
        X_train, X_test, y_train, y_test = train_test_split(train, y, 
                                                            random_state=random_state,
                                                            test_size=0.2)
        # 모델 학습
        model.fit(X_train, y_train)

        # 예측
        y_pred = model.predict(X_test)

        # 예측 결과의 rmse 값 저장
        df[model_name] = rmse(y_test, y_pred)

        # dataframe에 저장
        score_df = pd.DataFrame(df, index=['RMSE']).T.sort_values('RMSE', ascending=False)
        
    return score_df

get_scores(models, train, y)
```

### Grid Search

- Random vs Grid Search
- https://shwksl101.github.io/ml/dl/2019/01/30/Hyper_parameter_optimization.html

Grid Search의 param
- param_grid : 탐색할 파라미터의 종류 (딕셔너리로 입력)
- scoring : 모델의 성능을 평가할 지표
    - 다양한 scoring 사용가능
    - https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
- cv : cross validation을 수행하기 위해 train 데이터셋을 나누는 조각의 개수
- verbose : 그리드 탐색을 진행하면서 진행 과정을 출력해서 보여줄 메세지의 양 (숫자가 클수록 더 많은 메세지를 출력합니다.)
- n_jobs : 그리드 탐색을 진행하면서 사용할 CPU의 개수

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
def grid_search(model, train, y, param_grid, verbose=2, n_jobs=5):
    grid_model = GridSearchCV(model, param_grid=param_grid,
                         scoring='neg_mean_squared_error',
                         cv=5, verbose=verbose, n_jobs=n_jobs)

    grid_model.fit(train,y)
    
    params = grid_model.cv_results_['params']
    score = grid_model.cv_results_['mean_test_score']
    
    results = pd.DataFrame(params)
    results['RMSLE'] = score
    
    results['RMSLE'] = np.sqrt(-results['RMSLE'])
    results = results.sort_values('RMSLE')
    
    return results
```

```python
def get_cv_score(models, train, y):
    kfold = KFold(n_splits=5, random_state=random_state).get_n_splits(train.values)
    for m in models:
        print("Model {} CV score : {:.4f}".format(
            m['name'], np.mean(cross_val_score(m['model'], train.values,  y)), kf=kfold))
```

```python
def AveragingBlending(models, x, y, sub_x, name=''):
    for m in models:
        m['model'].fit(x.values, y)
    
    predictions = np.column_stack([
        m['model'].predict(sub_x.values) for m in models
    ])
    
    y_pred = np.mean(predictions, axis=1)
    y_pred = np.expm1(y_pred)
    
    submission_path = join(data_dir, 'sample_submission.csv')
    submission = pd.read_csv(submission_path)
    submission['price'] = y_pred

    submission_csv_path = '{}/submission_{}_{}.csv'.format(data_dir, 'avg_blending', name)
    submission.to_csv(submission_csv_path, index=False)
    print(submission_csv_path)
```
