### full code

```python
class Employee:
    def __init__ (self, name, division, jobgrade, years, basic):
        self.name = name
        self.division = division
        self.jobgrade = int(jobgrade)
        self.years = int(years)
        self.basic = float(basic)
        self.salary = 0

    def get_name(self):
        return self.name

    def get_divison(self):
        return self.division
    
    def get_jobgrade(self):
        return self.jobgrade
    
    def get_years(self):
        return self.years
    
    def get_basic(self):
        return self.basic

    def get_salary(self):
        self.salary = self.basic + (self.basic * self.years * 0.1)
        return self.salary

def make_employee(input_data):
    name, division, job, year, basic = input_data.split(',')
    return Employee(name, division, job, year, basic)

def summarize():
    file_path = '/home/aiffel0049/aiffel/Foundamental/15.database/data/employeefile.txt'
    input_file = open(file_path, 'r')
    highest = make_employee(input_file.readline())
    for line in input_file:
        e = make_employee(line)
        if e.get_salary() > highest.get_salary():
            highest = e
    input_file.close()
    
    print('Highest employee : ', highest.get_name())
    print('Years of service : ', highest.get_years())
    print('Division : ', highest.get_divison())
    print('Salary : ', highest.get_salary())
```

```python
summarize()
```

#### pd.merge

```python
import pandas as pd

df1 = pd.DataFrame({'Student': ['KimTaemin','HaJaehwa','JungSayoung','Sonjimin','Leesoomin','KangJun'],
                    'Korean': [90, 85, 88, 35, 40, 44],
                    'English': [80, 90, 40, 44, 55, 90]})
df2 = pd.DataFrame({'Student': ['KimTaemin','HaJaehwa','JungSayoung','Sonjimin','Leesoomin','KangJun'],
                    'Math': [100, 55, 38, 43, 68, 82]})

print(df1)
print('---')
print(df2)
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='Student')
```

- on= 인자값에 key설정 가능

```python
df1 = pd.DataFrame({'Student': ['KimTaemin','HaJaehwa','JungSayoung','Sonjimin','Leesoomin','KangJun'],
                    'Korean': [90, 85, 88, 35, 40, 44],
                    'English': [80, 90, 40, 44, 55, 90]})
df2 = pd.DataFrame({'Student': ['Jiyoungmin','KimTaemin'],
                    'Math':[44,33]})

print(df1)
print(df2)
```
