## 이미지 로드

```python
from PIL import Image, ImageColor
import os
img_path = os.getcwd() + "/newyork.jpg"
img = Image.open(img_path)
print(img_path)
print(type(img))
img
```

## 이미지의 속성(크기, 압축형태, 색 표현모드)

```python
W, H = img.size
print((W, H))
```

```python
print(img.format)
print(img.size)
print(img.mode)
```

## Cropping image

```python
img.crop((30,30,100,100))
```

```python
# 새로운 이미지 파일명
cropped_img_path = os.getcwd() + "/croped_img.jpg"
img.crop((30,30,100,100)).save(cropped_img_path)
```

## image ndarray로 변환

```python
import numpy as np
img_arr = np.array(img)
print(type(img))
print(type(img_arr))
print(img_arr.shape)
print(img_arr.ndim)
```

```python
img_arr
```

## 이미지 Gray scale로 변환

```python
img_g = Image.open(img_path).convert('L')
img_g
```

```python
img_g_arr = np.array(img_g)
print(type(img_g_arr))
print(img_g_arr.shape)
print(img_g_arr.ndim)
```

```python
img_g_arr
```

## 원하는 색 데이터 가져오기

```python
red = ImageColor.getcolor('RED','RGB')
reda = ImageColor.getcolor('red','RGBA')
yellow = ImageColor.getcolor('yellow','RGB')
print(red)
print(reda)
print(yellow)
```
